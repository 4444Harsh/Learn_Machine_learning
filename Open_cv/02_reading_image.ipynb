```python
# r1=length,r2=width,r3=channel
img_cv.shape
```

```python
# numpy slicnig
#BGR
b=img_cv[:,:,0]
g=img_cv[:,:,1]
r=img_cv[:,:,2]
```

```python
cv2.imshow('blue channel',b)
cv2.imshow('green channel',g)
cv2.imshow('red channel',r)

cv2.imshow('BGR',img_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#opencv split channel
b,g,r=cv2.split(img_cv)
```

```python
cv2.imshow('blue channel',b)
cv2.imshow('green channel',g)
cv2.imshow('red channel',r)

cv2.imshow('BGR',img_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
